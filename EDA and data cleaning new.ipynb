```python
#filling in missing values

house_data.isnull().sum()
```

```python
# filling the address column with the parcelid column because each parcelid is distinct to the address

house_data['PropertyAddress'] = house_data.groupby('ParcelID')['PropertyAddress'].transform(lambda x: x.fillna
                                                                                            (x.mode().iloc[0]))

# this function groups the df based on the parcelid and the transform function is applied to each group, it uses the
# the lambda function to fill the missing values in the address column(x), with the mode(most frequent) of the 
# group's address column. the iloc[0] part retrives the 1st mode incase there are multiple
```

```python
house_data['OwnerName'] = house_data['OwnerName'].fillna('not provided')
```

```python
house_data['OwnerAddress'] = house_data['OwnerAddress'].fillna('not provided')
```

```python
house_data.hist(figsize=(15,8))
```

```python
house_data['Acreage'] = house_data.groupby('LandUse')['Acreage'].transform(lambda x: x.fillna(x.median()))
```

```python
def fill_missing_median(group):
    if group.notna().any():
        median_value = group.median()
        return group.fillna(median_value)
    else:
        return group
```

```python
def fill_missing_mean(group):
    if group.notna().any():
        mean_value = group.mean()
        return group.fillna(round(mean_value))
    else:
        return group
```

```python
house_data['LandValue'] = house_data.groupby('LandUse')['LandValue'].transform(fill_missing_median)
```

```python
house_data['BuildingValue'] = house_data.groupby('LandUse')['BuildingValue'].transform(fill_missing_median)
```

```python
house_data['TotalValue'] = house_data.groupby('LandUse')['TotalValue'].transform(fill_missing_median)
```

```python
house_data['Bedrooms'] = house_data.groupby('LandUse')['Bedrooms'].transform(fill_missing_mean)
```

```python
house_data['FullBath'] = house_data.groupby('LandUse')['FullBath'].transform(fill_missing_mean)
```

```python
house_data['HalfBath'] = house_data.groupby('LandUse')['HalfBath'].transform(fill_missing_median)
```
