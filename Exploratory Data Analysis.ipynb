```python
# looking at countries with the least population

least_pop = world_pop.sort_values(by='2022 Population').head()
least_pop[['2022 Population', 'Country']].plot(kind='bar', x='Country', y='2022 Population', figsize=(8,6), legend=False)
plt.title('Countries with the least population')
```

```python
# looking at countries with the highest population

high_pop = world_pop.sort_values(by='2022 Population', ascending=False).head()

plt.figure(figsize=(8,6))
ax = sns.barplot(x='Country', y='2022 Population', data=high_pop)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Countries with the highest population')
```

```python
# looking at countries that have the highest world%

world_per = world_pop.sort_values(by='World Population Percentage', ascending=False).head()
ax = sns.barplot(x='Country', y='World Population Percentage', data=world_per)
                                                           

for bar in ax.containers:
    ax.bar_label(bars)
    
plt.title('Top 5 Countries with the highest world %')
plt.show()
```
